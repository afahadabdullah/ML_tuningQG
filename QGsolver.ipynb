```python
#!/usr/bin/env python
# coding: utf-8
"""
QG two‑layer mini‑solver (no pyqg) with **shared physics and shared ICs** for
high‑res ("truth") and low‑res ("model"). The **only intentional difference** is
spatial/temporal resolution (nx, dt). Everything else is identical by default.

This version adds stability + anti‑noise fixes so both runs evolve coherently:
- 2/3‑rule **de‑aliasing** for the nonlinear term (on by default)
- **AB2 + Robert–Asselin (RA) filter** for time stepping
- Optional **Helmholtz (α‑model) filter** for velocities used in advection
- Sensible, large‑scale **ring forcing** (default kf=8) to avoid pixel noise
- ν₄ scaled by Δx^p **identically** across resolutions (default p=4)

Figures & animations use **cmap='PuOr'**.

**New (for propagation):** shared β‑plane and uniform mean zonal flow `Umean` so patterns drift **left→right**. The PV equations now include `+ β ∂ψ/∂x` and mean‑flow advection `+ Umean ∂q/∂x` as linear spectral terms.
"""
from __future__ import annotations

import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from dataclasses import dataclass
from tqdm import trange

# ------------------------- utils -------------------------

def coarse_grain(field: np.ndarray, factor: int) -> np.ndarray:
    ny, nx = field.shape
    ny_c, nx_c = ny // factor, nx // factor
    return field[:ny_c*factor, :nx_c*factor].reshape(ny_c, factor, nx_c, factor).mean(axis=(1, 3))

# ------------------------- config ------------------------
@dataclass
class Config:
    # Mode / grid / physics (shared)
    run_mode: str = "truth"        # 'truth' | 'model'
    L: float = 1_000e3             # [m]
    nx: int = 256                  # overridden in setup_resolution
    dt: float = 900.0              # overridden in setup_resolution
    days: float = 30.0

    rd: float = 40e3               # deformation radius [m]
    rek: float = 3.0e-8            # bottom drag [s^-1]

    # Planetary vorticity + mean flow (shared)
    beta: float = 2.0e-11          # meridional PV gradient [1/(m·s)]
    Umean: float = 0.30            # uniform eastward mean flow [m/s]

    # Dissipation — shared & scaled identically by Δx^p
    nu4_ref: float = 5e9
    scale_nu4_by_dx: bool = True
    nu4_scale_power: float = 4.0   # p in ν4 ∝ (Δx)^p (same for both)

    # Forcing — shared
    forcing_type: str = 'ring'
    forcing_kf: float = 8.0        # larger‑scale injection by default (was 16)
    forcing_bw: float = 2.0
    forcing_tau: float = 86400.0   # s
    forcing_power: float = 5e-17

    # Numerics — shared
    scheme: str = 'ab2'            # AB2 with RA filter
    ra_epsilon: float = 0.05       # Robert–Asselin coefficient (0.02–0.1 typical)
    dealias_23: bool = True        # enable 2/3‑rule de‑aliasing
    helmholtz_alpha_over_dx: float = 1.5  # α/Δx for velocity filtering (0 disables)

    snap_every: int = 8
    out_nc: str = 'qg_out.nc'

    # IC I/O (truth saves; model loads & coarse‑grains)
    save_ic_path: str | None = None
    load_ic_path: str | None = None

# -------------------- resolution presets -----------------

def setup_resolution(cfg: Config) -> Config:
    if cfg.run_mode == 'truth':
        cfg.nx = 256; cfg.dt = 600.0
        cfg.out_nc = 'qg_truth_256.nc'
        cfg.save_ic_path = 'ic_truth_highres.npz'
        cfg.load_ic_path = None
    elif cfg.run_mode == 'model':
        cfg.nx = 64; cfg.dt = 1200.0   # 20‑min for better stability
        cfg.out_nc = 'qg_model_64.nc'
        cfg.save_ic_path = None
        cfg.load_ic_path = 'ic_truth_highres.npz'
    else:
        raise ValueError("run_mode must be 'truth' or 'model'")

    dx = cfg.L / cfg.nx
    dx_ref = cfg.L / 256.0
    nu4 = cfg.nu4_ref * ((dx / dx_ref) ** cfg.nu4_scale_power) if cfg.scale_nu4_by_dx else cfg.nu4_ref

    print("="*72)
    print(f"{cfg.run_mode.upper()} CONFIG: nx={cfg.nx}, dt={cfg.dt/60:.1f} min, nu4={nu4:.2e}, rek={cfg.rek:.1e}")
    print(f"Forcing: {cfg.forcing_type} (kf={cfg.forcing_kf}, bw={cfg.forcing_bw}, power={cfg.forcing_power:.1e}, tau={cfg.forcing_tau/86400:.2f} d)")
    print(f"Scheme: {cfg.scheme} | RA ε={cfg.ra_epsilon:.3f} | de‑alias={cfg.dealias_23} | α/Δx={cfg.helmholtz_alpha_over_dx}")
    print(f"β={cfg.beta:.2e} 1/(m·s), Umean={cfg.Umean:.2f} m/s (eastward)")

    cfg.nu4 = float(nu4); cfg.dx = float(dx)
    return cfg

# ---------------- spectral operators + state --------------

def initialize_model(cfg: Config):
    nx, L, dx = cfg.nx, cfg.L, cfg.dx

    kx = np.fft.rfftfreq(nx, d=dx) * 2*np.pi
    ky = np.fft.fftfreq(nx, d=dx) * 2*np.pi
    KX, KY = np.meshgrid(kx, ky)
    K2 = KX**2 + KY**2
    K = np.sqrt(K2)
    F = (1.0/cfg.rd)**2

    dissipation_factor = np.exp(-cfg.nu4 * K2**2 * cfg.dt)
    drag_factor = np.exp(-cfg.rek * cfg.dt)

    # 2/3‑rule mask (rectangular in kx, ky)
    kx_max = kx.max(); ky_max = np.abs(ky).max()
    mask_23 = (np.abs(KX) <= (2.0/3.0)*kx_max) & (np.abs(KY) <= (2.0/3.0)*ky_max)

    print(f"[INIT] Grid: {nx}x{nx}, dx={dx/1e3:.1f} km; Rd={cfg.rd/1e3:.0f} km")
    print(f"[INIT] Dissipation at k_max: {dissipation_factor.min():.4f}; drag: {drag_factor:.6f}")

    # IC (shared)
    if cfg.load_ic_path:
        data = np.load(cfg.load_ic_path)
        psi1_hi, psi2_hi = data['psi1'], data['psi2']
        factor = psi1_hi.shape[0] // nx
        psi1 = coarse_grain(psi1_hi, factor)
        psi2 = coarse_grain(psi2_hi, factor)
        print(f"[IC] Loaded {cfg.load_ic_path} and coarse‑grained by {factor}× → {nx}²")
    else:
        y = np.linspace(0, L, nx, endpoint=False)
        x = np.linspace(0, L, nx, endpoint=False)
        X, Y = np.meshgrid(x, y)
        sigma = 0.15 * L
        psi1 = 10.0*(np.exp(-((X-0.3*L)**2 + (Y-0.5*L)**2)/(2*sigma**2)) -
                     np.exp(-((X-0.7*L)**2 + (Y-0.5*L)**2)/(2*sigma**2)))
        psi2 = -0.6*psi1
        if cfg.save_ic_path:
            np.savez(cfg.save_ic_path, psi1=psi1, psi2=psi2)
            print(f"[IC] Saved initial ψ fields → {cfg.save_ic_path}")

    psi1_h = np.fft.rfft2(psi1)
    psi2_h = np.fft.rfft2(psi2)
    q1_h = -K2*psi1_h + F*(psi2_h - psi1_h)
    q2_h = -K2*psi2_h + F*(psi1_h - psi2_h)

    # forcing mask (shared)
    def cycles_to_wavenumber(c):
        return 2*np.pi*c / L
    forcing_mask = np.zeros_like(K)
    if cfg.forcing_type == 'ring':
        kf = cycles_to_wavenumber(cfg.forcing_kf)
        bw = cycles_to_wavenumber(cfg.forcing_bw)
        forcing_mask = np.exp(-((K - kf)**2)/(2*bw**2)); forcing_mask[K==0] = 0.0

    Fh1 = np.zeros_like(q1_h, dtype=np.complex128)
    Fh2 = np.zeros_like(q2_h, dtype=np.complex128)

    state = dict(
        q1_h=q1_h, q2_h=q2_h,
        KX=KX, KY=KY, K2=K2, K=K, F=F,
        dissipation_factor=dissipation_factor, drag_factor=drag_factor,
        dx=dx,
        mask_23=mask_23,
        forcing_mask=forcing_mask,
        Fh1=Fh1, Fh2=Fh2,
        adv1_prev=None, adv2_prev=None,
        q1_prev=None, q2_prev=None,
        beta=float(cfg.beta), Umean=float(cfg.Umean),
    )
    return state

# --------------- inversion, advection, forcing -------------

def pv_to_streamfunction(q1_h, q2_h, K2, F):
    K2s = K2.copy(); K2s[0,0] = 1.0
    det = (K2s+F)**2 - F**2
    psi1_h = (-(K2s+F)*q1_h + F*q2_h)/det
    psi2_h = ( F*q1_h - (K2s+F)*q2_h)/det
    psi1_h[0,0] = 0.0; psi2_h[0,0] = 0.0
    return psi1_h, psi2_h


def helmholtz_filter(psi_h, K2, alpha_over_dx, dx):
    if alpha_over_dx <= 0.0:
        return psi_h
    alpha = alpha_over_dx * dx
    return psi_h / (1.0 + (alpha**2) * K2)


def compute_advection(q_h, psi_h, KX, KY, cfg: Config, state):
    # optional pre‑product de‑aliasing
    if cfg.dealias_23:
        m = state['mask_23']
        psi_h = psi_h * m
        q_h   = q_h   * m
    # optional Helmholtz filter for velocities
    psi_h = helmholtz_filter(psi_h, state['K2'], cfg.helmholtz_alpha_over_dx, state['dx'])

    u_h =  1j * KY * psi_h
    v_h = -1j * KX * psi_h
    dqdx_h = 1j * KX * q_h
    dqdy_h = 1j * KY * q_h

    u = np.fft.irfft2(u_h).real; v = np.fft.irfft2(v_h).real
    dqdx = np.fft.irfft2(dqdx_h).real; dqdy = np.fft.irfft2(dqdy_h).real

    adv = -(u * dqdx + v * dqdy)
    adv_h = np.fft.rfft2(adv)

    if cfg.dealias_23:
        adv_h *= state['mask_23']
    return adv_h


def update_ou(Fh, mask, dt, tau, power):
    if tau <= 0 or np.all(mask == 0):
        return Fh
    a = np.exp(-dt / tau)
    b = np.sqrt((1 - a**2) * power)
    noise = (np.random.normal(size=Fh.shape) + 1j*np.random.normal(size=Fh.shape))
    return a * Fh + b * mask * noise

# -------------------------- stepping ----------------------

def time_step(state, cfg: Config):
    q1_h, q2_h = state['q1_h'], state['q2_h']
    KX, KY, K2, F = state['KX'], state['KY'], state['K2'], state['F']
    dt = cfg.dt

    psi1_h, psi2_h = pv_to_streamfunction(q1_h, q2_h, K2, F)

    adv1 = compute_advection(q1_h, psi1_h, KX, KY, cfg, state)
    adv2 = compute_advection(q2_h, psi2_h, KX, KY, cfg, state)

    if cfg.scheme.lower() == 'ab2' and state['adv1_prev'] is not None:
        adv1 = 1.5*adv1 - 0.5*state['adv1_prev']
        adv2 = 1.5*adv2 - 0.5*state['adv2_prev']
    state['adv1_prev'], state['adv2_prev'] = adv1.copy(), adv2.copy()

    # β-term:  +β ∂ψ/∂x  ⇒ spectral term  -i kx β ψ̂
    beta1 = -1j * KX * psi1_h * state['beta']
    beta2 = -1j * KX * psi2_h * state['beta']
    
    # Mean-flow advection: +Umean ∂q/∂x ⇒ spectral term -i kx U q̂
    flow1 = -1j * KX * state['Umean'] * q1_h
    flow2 = -1j * KX * state['Umean'] * q2_h

    # Forcing (shared)
    if cfg.forcing_type == 'ring':
        state['Fh1'] = update_ou(state['Fh1'], state['forcing_mask'], dt, cfg.forcing_tau, cfg.forcing_power)
        state['Fh2'] = update_ou(state['Fh2'], state['forcing_mask'], dt, cfg.forcing_tau, cfg.forcing_power)
        Fr1, Fr2 = state['Fh1'], state['Fh2']
    else:
        Fr1 = 0.0; Fr2 = 0.0

    # Save previous (for RA filter)
    q1_prev = state['q1_prev']; q2_prev = state['q2_prev']
    q1_new = q1_h + dt * (adv1 + beta1 + flow1 + Fr1)
    q2_new = q2_h + dt * (adv2 + beta2 + flow2 + Fr2)

    # Hyperviscosity + drag
    q1_new *= state['dissipation_factor']
    q2_new *= state['dissipation_factor'] * state['drag_factor']

    # Robert–Asselin filter (acts on the leapfrog family, used widely with AB2 too)
    eps = float(cfg.ra_epsilon)
    if eps > 0.0 and (q1_prev is not None):
        q1_new = q1_new + eps * (q1_h - 2.0*q1_new + q1_prev)
        q2_new = q2_new + eps * (q2_h - 2.0*q2_new + q2_prev)

    # update state
    state['q1_prev'], state['q2_prev'] = q1_h, q2_h
    state['q1_h'], state['q2_h'] = q1_new, q2_new
    return state

# --------------------------- driver -----------------------

def run_simulation(cfg: Config):
    cfg = setup_resolution(cfg)
    state = initialize_model(cfg)
    total_steps = int(cfg.days * 86400 / cfg.dt)

    times, psi_bt_list, zeta1_list, psi1_list = [], [], [], []

    for step in trange(total_steps, desc=f"QG {cfg.run_mode}"):
        state = time_step(state, cfg)
        if (step + 1) % cfg.snap_every == 0:
            psi1_h, psi2_h = pv_to_streamfunction(state['q1_h'], state['q2_h'], state['K2'], state['F'])
            psi1 = np.fft.irfft2(psi1_h, s=(cfg.nx, cfg.nx)).real
            psi2 = np.fft.irfft2(psi2_h, s=(cfg.nx, cfg.nx)).real
            psi_bt = 0.5*(psi1 + psi2)
            zeta1 = -np.fft.irfft2(state['K2'] * psi1_h, s=(cfg.nx, cfg.nx)).real
            times.append((step+1)*cfg.dt)
            psi_bt_list.append(psi_bt.astype(np.float32))
            zeta1_list.append(zeta1.astype(np.float32))
            psi1_list.append(psi1.astype(np.float32))
            if len(times) % 20 == 0:
                print(f"  t={times[-1]/86400:.1f} d | max|ζ|={np.abs(zeta1).max():.2e}")

    ds = xr.Dataset(
        {
            'psi_bt': (('time','y','x'), np.stack(psi_bt_list)),
            'psi1':   (('time','y','x'), np.stack(psi1_list)),
            'zeta1':  (('time','y','x'), np.stack(zeta1_list)),
        },
        coords=dict(
            time=np.array(times)/86400.0,
            y=np.linspace(0, cfg.L, cfg.nx, endpoint=False),
            x=np.linspace(0, cfg.L, cfg.nx, endpoint=False),
        ),
        attrs={'L': float(cfg.L), 'nx': int(cfg.nx), 'nu4': float(cfg.nu4), 'rek': float(cfg.rek), 'dt': float(cfg.dt),
               'scheme': str(cfg.scheme), 'forcing': str(cfg.forcing_type)}
    )
    ds.to_netcdf(cfg.out_nc)
    print(f"[OK] Saved {cfg.out_nc} with {len(times)} snapshots")
    return ds

# ---------------------------- compare ---------------------

def compare_statistics(ds_truth, ds_model, last_days: float | None = None):
    print("" + "="*72)
    print("STATISTICAL COMPARISON")
    print("="*72)
    factor = ds_truth.sizes['x'] // ds_model.sizes['x']
    assert ds_truth.sizes['x'] % ds_model.sizes['x'] == 0

    # align by common physical times
    t_truth_sec = (ds_truth['time'].values * 86400.0).astype(float)
    t_model_sec = (ds_model['time'].values * 86400.0).astype(float)
    t_truth_ri = np.round(t_truth_sec).astype(int)
    t_model_ri = np.round(t_model_sec).astype(int)
    t_common = np.intersect1d(t_truth_ri, t_model_ri)
    if last_days is not None and last_days > 0:
        end_sec = t_common.max(); start_sec = end_sec - int(last_days*86400)
        t_common = t_common[(t_common >= start_sec) & (t_common <= end_sec)]
    sel_truth = np.isin(t_truth_ri, t_common); sel_model = np.isin(t_model_ri, t_common)

    zt = ds_truth['zeta1'].values[sel_truth]; pt = ds_truth['psi_bt'].values[sel_truth]
    zm = ds_model['zeta1'].values[sel_model]; pm = ds_model['psi_bt'].values[sel_model]

    zt_cg = np.stack([coarse_grain(zt[i], factor) for i in range(zt.shape[0])])
    pt_cg = np.stack([coarse_grain(pt[i], factor) for i in range(pt.shape[0])])

    print(f"Aligned common frames: {zt_cg.shape[0]} | coarse‑grain factor {factor}x")
    rms_z = np.sqrt(((zm - zt_cg)**2).mean()); corr_z = np.corrcoef(zm.ravel(), zt_cg.ravel())[0,1]
    rms_p = np.sqrt(((pm - pt_cg)**2).mean()); corr_p = np.corrcoef(pm.ravel(), pt_cg.ravel())[0,1]
    print(f"RMS(ζ): {rms_z:.3e}, Corr(ζ): {corr_z:.3f}")
    print(f"RMS(ψ): {rms_p:.3e}, Corr(ψ): {corr_p:.3f}")
    return zt_cg, zm, pt_cg, pm

# --------------------------- plotting ---------------------

def _energy_proxy_from_psi(psi_series: np.ndarray, dx: float) -> np.ndarray:
    ke = []
    for psi in psi_series:
        dpsy, dpsx = np.gradient(psi, dx, dx, edge_order=2)
        ke.append(0.5*np.mean(dpsx**2 + dpsy**2))
    return np.array(ke)


def make_comparison_figure(zt_cg, zm, pt_cg, pm, dx, save_path='qg_detailed_comparison.png'):
    z_t = zt_cg[-1]; z_m = zm[-1]
    p_t = pt_cg[-1]; p_m = pm[-1]

    zbins = np.linspace(min(z_t.min(), z_m.min()), max(z_t.max(), z_m.max()), 60)
    pbins = np.linspace(min(p_t.min(), p_m.min()), max(p_t.max(), p_m.max()), 60)

    qt = np.sort(z_t.ravel()); qm = np.sort(z_m.ravel())
    n = min(qt.size, qm.size); qt = qt[:n]; qm = qm[:n]

    maxz_t = np.max(np.abs(zt_cg), axis=(1,2)); maxz_m = np.max(np.abs(zm), axis=(1,2))
    ke_t = _energy_proxy_from_psi(pt_cg, dx); ke_m = _energy_proxy_from_psi(pm, dx)
    ratio = (np.std(zm, axis=(1,2)) / (np.std(zt_cg, axis=(1,2)) + 1e-12))

    fig = plt.figure(figsize=(16,10))
    ax = plt.subplot2grid((3,3),(0,0)); im0=ax.imshow(z_t,origin='lower',cmap='PuOr'); ax.set_title('Truth (coarse)'); plt.colorbar(im0,ax=ax,fraction=0.046)
    ax = plt.subplot2grid((3,3),(0,1)); im1=ax.imshow(z_m,origin='lower',cmap='PuOr'); ax.set_title('Model'); plt.colorbar(im1,ax=ax,fraction=0.046)
    ax = plt.subplot2grid((3,3),(0,2)); im2=ax.imshow(z_m-z_t,origin='lower',cmap='PuOr'); ax.set_title('Difference (model − truth)'); plt.colorbar(im2,ax=ax,fraction=0.046)

    ax = plt.subplot2grid((3,3),(1,0))
    ax.hist(z_t.ravel(),bins=zbins,density=True,alpha=0.6,label='Truth')
    ax.hist(z_m.ravel(),bins=zbins,density=True,alpha=0.6,label='Model')
    ax.set_title('Vorticity Distribution'); ax.set_xlabel('ζ'); ax.set_ylabel('PDF'); ax.set_yscale('log'); ax.legend()

    ax = plt.subplot2grid((3,3),(1,1))
    ax.hist(p_t.ravel(),bins=pbins,density=True,alpha=0.6,label='Truth')
    ax.hist(p_m.ravel(),bins=pbins,density=True,alpha=0.6,label='Model')
    ax.set_title('Streamfunction Distribution'); ax.set_xlabel('ψ'); ax.set_ylabel('PDF'); ax.legend()

    ax = plt.subplot2grid((3,3),(1,2))
    lo = min(qt.min(), qm.min()); hi = max(qt.max(), qm.max())
    ax.plot(qt,qm,'.',ms=2); ax.plot([lo,hi],[lo,hi],'--'); ax.set_title('Q–Q Plot (ζ)'); ax.set_xlabel('Truth quantiles'); ax.set_ylabel('Model quantiles')

    t = np.arange(len(maxz_t))
    ax = plt.subplot2grid((3,3),(2,0)); ax.plot(t,maxz_t,label='Truth'); ax.plot(t,maxz_m,label='Model'); ax.set_title('Vorticity Evolution'); ax.set_xlabel('Frame'); ax.set_ylabel('max|ζ|'); ax.legend()
    ax = plt.subplot2grid((3,3),(2,1)); ax.plot(t,ke_t,label='Truth'); ax.plot(t,ke_m,label='Model'); ax.set_title('Energy Evolution'); ax.set_xlabel('Frame'); ax.set_ylabel('KE proxy'); ax.legend()
    ax = plt.subplot2grid((3,3),(2,2)); ax.plot(t,ratio); ax.axhline(1.0, linestyle='--'); ax.fill_between(t,0.9,1.1,alpha=0.2,label='Acceptable'); ax.set_title('Vorticity Ratio Evolution'); ax.set_xlabel('Frame'); ax.set_ylabel('Ratio'); ax.legend()

    fig.tight_layout(); fig.savefig(save_path,dpi=200); print(f"[OK] Saved comparison figure → {save_path}")

# ----------------------- animation ------------------------

def make_side_by_side_animation_from_nc(truth_nc='qg_truth_256.nc', model_nc='qg_model_64.nc', var='psi1', sample_hours=6, frame_skip=1, out_path='qg_anim_psi1_from_ic.gif', fps=10, include_ic=True):
    dst = xr.open_dataset(truth_nc); dsm = xr.open_dataset(model_nc)
    t_t = (dst['time'].values*86400.0).astype(float); t_m = (dsm['time'].values*86400.0).astype(float)
    start = max(t_t.min(), t_m.min()); end = min(t_t.max(), t_m.max()); step = float(sample_hours)*3600.0
    grid0 = np.ceil(start/step)*step; gridN = np.floor(end/step)*step
    if gridN < grid0: raise ValueError('No overlapping window to sample for animation.')
    grid = np.arange(grid0, gridN+0.5*step, step)

    def nearest_indices(tarr, G):
        idx = np.searchsorted(tarr, G); idx = np.clip(idx,1,len(tarr)-1)
        left = idx-1; right = idx
        return np.where(np.abs(tarr[right]-G) < np.abs(tarr[left]-G), right, left)

    it = nearest_indices(t_t, grid); im = nearest_indices(t_m, grid)
    At = dst[var].values[it]; Am = dsm[var].values[im]

    factor = dst.sizes['x']//dsm.sizes['x']; assert dst.sizes['x'] % dsm.sizes['x'] == 0
    At_cg = np.stack([coarse_grain(At[i], factor) for i in range(At.shape[0])])

    labels_days = grid/86400.0
    if include_ic and var in ('psi_bt','psi1'):
        nx_m = dsm.sizes['x']; data = np.load('ic_truth_highres.npz')
        psi1_hi, psi2_hi = data['psi1'], data['psi2']
        factor_ic = psi1_hi.shape[0]//nx_m
        psi1_ic = coarse_grain(psi1_hi, factor_ic); psi2_ic = coarse_grain(psi2_hi, factor_ic)
        psi_ic_bt = 0.5*(psi1_ic + psi2_ic); psi_ic = psi_ic_bt if var=='psi_bt' else psi1_ic
        At_cg = np.concatenate([psi_ic[None,...], At_cg], axis=0)
        Am = np.concatenate([psi_ic[None,...], Am], axis=0)
        labels_days = np.concatenate([[0.0], labels_days])

    sel = np.arange(At_cg.shape[0])[::max(1,int(frame_skip))]
    At_cg = At_cg[sel]; Am = Am[sel]; labels_days = labels_days[sel]

    # moving color limits per‑frame based on high‑res but **shared** across panels
    vmins = np.minimum(At_cg.min(axis=(1,2)), Am.min(axis=(1,2)))
    vmaxs = np.maximum(At_cg.max(axis=(1,2)), Am.max(axis=(1,2)))

    fig, axes = plt.subplots(1,2,figsize=(10,5),constrained_layout=True)
    im0 = axes[0].imshow(At_cg[0], origin='lower', cmap='PuOr'); im1 = axes[1].imshow(Am[0], origin='lower', cmap='PuOr')
    title = fig.suptitle(f'Day {labels_days[0]:.2f}')
    axes[0].set_title('Truth (coarse) — ψ₁'); axes[1].set_title('Model — ψ₁')
    cbar = fig.colorbar(im1, ax=axes.ravel().tolist(), fraction=0.025, pad=0.02); cbar.ax.set_ylabel(var)

    def _update(i):
        im0.set_data(At_cg[i]); im1.set_data(Am[i])
        im0.set_clim(vmins[i], vmaxs[i]); im1.set_clim(vmins[i], vmaxs[i])
        title.set_text(f'Day {labels_days[i]:.2f}')
        return im0, im1, title

    anim = animation.FuncAnimation(fig, _update, frames=At_cg.shape[0], interval=120, blit=True)
    try:
        from matplotlib.animation import PillowWriter
        anim.save(out_path, writer=PillowWriter(fps=fps))
        print(f"[OK] Saved animation → {out_path} (frames={At_cg.shape[0]})")
    except Exception as e:
        print(f"[WARN] GIF save failed ({e}); attempting MP4 …")
        try:
            anim.save(out_path.replace('.gif','.mp4'), writer='ffmpeg', fps=fps, dpi=150)
            print(f"[OK] Saved MP4 → {out_path.replace('.gif','.mp4')}")
        except Exception as e2:
            print(f"[FAIL] Could not save animation: {e2}")

# ----------------------------- main -----------------------
if __name__ == '__main__':
    # Shared physics; only nx/dt differ
    cfg_truth = Config(run_mode='truth', days=30)
    ds_truth = run_simulation(cfg_truth)

    cfg_model = Config(run_mode='model', days=30)
    ds_model = run_simulation(cfg_model)

    zt_cg, zm, pt_cg, pm = compare_statistics(ds_truth, ds_model, last_days=10.0)
    make_comparison_figure(zt_cg, zm, pt_cg, pm, dx=cfg_model.L/cfg_model.nx, save_path='qg_detailed_comparison.png')
    make_side_by_side_animation_from_nc(var='psi1', sample_hours=6, frame_skip=1, out_path='qg_anim_psi1_from_ic.gif', fps=10, include_ic=True)
```
